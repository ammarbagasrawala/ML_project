```python
from sklearn.model_selection import train_test_split ,RandomizedSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score ,confusion_matrix, precision_score, recall_score, ConfusionMatrixDisplay
from scipy.stats import randint
from sklearn.datasets import make_classification
from sklearn.model_selection import GridSearchCV

```

```python
x=df.drop('target',axis=1)
y=df["target"]
```

```python
# x=x.drop(["fbs"],axis=1)
x=x.drop(["chol"],axis=1)
```

```python

plt.figure(figsize=(16,10))
corr_matrix=df.corr()
sns.heatmap(corr_matrix,annot=True,cmap='coolwarm')
```

```python
x_train,x_test,y_train,y_test=train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
rfc_model=RandomForestClassifier(n_estimators=100,criterion="log_loss",bootstrap=True)
rfc_model.fit(x_train,y_train)
```

```python
y_pred=rfc_model.predict(x_test)
```

```python
acc=accuracy_score(y_test,y_pred)
acc
```

```python
rfc=RandomForestClassifier()
```

```python
params={
    'n_estimators': [10, 50, 100, 200],
    'max_depth': [None, 5, 10, 20],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['sqrt', 'log2', None]

}
```

```python
grid_search = GridSearchCV(rfc, param_grid=params, cv=5, n_jobs=-1)
grid_search.fit(x_train,y_train)
```

```python
print("Best hyperparameters:", grid_search.best_params_)
print("Best score:", grid_search.best_score_)
```

```python
best_hyperparameters={'max_depth': 5, 'max_features': 'sqrt', 'min_samples_leaf': 2, 'min_samples_split': 5, 'n_estimators': 50}
```

```python
rfc=RandomForestClassifier(**best_hyperparameters)
rfc.fit(x_train,y_train)
```
